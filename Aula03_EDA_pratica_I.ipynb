```python
import numpy as np
import scipy
from matplotlib import pyplot as plt

x = scipy.stats.norm.rvs(size=100)
y = scipy.stats.norm.rvs(size=100)
#data = np.array(list(zip(x,y)))
r = np.corrcoef(x,y)[1,0]


plt.scatter(x,y, color='b')
plt.plot(np.linspace(x.min(), x.max(), 100), np.linspace(x.min(), x.max(), 100)*r, color='r')
plt.title('r=%f'%r)
plt.show()


y.put(x.argmax(), 100)
r = np.corrcoef(x,y)[1,0]
plt.subplot(121)
plt.scatter(x,y, color='b')
plt.plot(np.linspace(x.min(), x.max(), 100), np.linspace(x.min(), x.max(), 100)*r, color='r')
plt.title('r=%f'%r)
plt.subplot(122)
ids = list(range(100))
i = x.argmax()
ids.pop(i)
plt.scatter(x[ids],y[ids], color='b')
plt.plot(np.linspace(x.min(), x.max(), 100), np.linspace(x.min(), x.max(), 100)*r, color='r')
plt.show()

# o primeiro gráfico é o gráfico de dispersão entre as variáveis x e y junto 
# com a correlação entre elas

# Para os dois gráficos de baixo foi modificado um ponto, o mais a direita
# o grafico a esquerda é o mesmo que a direita, mas como um ponto tem valor Y 100 fica ruim visualizar
# no grafico a direita não está sendo exibido esse ponto com valor 100
# apenas para ficar mais fácil compara com o gráfico de cima

# perceba como a correlação mudou

# como os dados são gerados aleatoriamente, pode ocorrer do primeiro gráfico já ter uma
# correlação alta, execute esse código algumas vezes para perceber isso
```

```python
# como tem 4 atributos o describe já calcula bastante coisa
print(iris.describe())

print('\nModa %s'%[iris[c].mode() for c in iris.columns])
# ruim de visualizar, fazer na mão para entender melhor

print('\nModa primeira coluna %s'%iris.iloc[:, [0]].mode())
print('Moda segunda coluna %s'%iris.iloc[:, [1]].mode())
print('Moda terceira coluna %s'%iris.iloc[:, [2]].mode())
print('Moda quarta coluna %s'%iris.iloc[:, [3]].mode())
```

```python
# quantos valores únicos de cada classe há ?

classes = iris.target.unique()
print('Mostrar as classes %s\n'%classes)
print('exemplos por classe %s'%[iris[iris.target == c].count() for c in classes])
# assim mostra o número de exemplos para cada coluna, não quero assim
print('\nexemplos por classe %s'%[iris[iris.target == c].count()[0] for c in classes])

# o número de exemplos por classe está bem distribuído
```

```python
# farei a análise acima por classe

for c in classes:
    print('\nClasse %s\n%s'%(c, iris[iris.target == c].describe()))
    
```

```python
# tem muita informação junta e está difícil comparar as coisas, 
# vou mostrar apenas a média e desvio padrão e mediana entre as classes

for c in classes:
    print('\nClasse %s'%c)
    sample = iris[iris.target == c]
    cols = sample.columns
    for col in cols[:-1]:
        print('atributo %s %.2f %.2f %.2f'%(col, sample[col].mean(), sample[col].std(), sample[col].median()))
```

```python
# tendo em mente as perguntas que fiz no ítem 1 ainda quero descobrir se há correlação entre
# algumas comprimento e largura

print('correlacao para sepata %s'%iris.iloc[:, [0,1]].corr())

print('correlacao para petala %s'%iris.iloc[:, [2,3]].corr())
```
